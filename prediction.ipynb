```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.models  import load_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder 
import pickle
```

```python
model=load_model('model.h5')
```

```python
with open ('label_encoder.pkl','rb') as f:
   label_encoder= pickle.load(f)
with open ('ohe_encoder.pkl','rb') as f:
   ohe_encoder= pickle.load(f)
with open ('scaler.pkl','rb') as f:
   scaler= pickle.load(f)   
```

```python
input_data= {
    'CreditScore': 600,
    'Gender':'Male',
    'Geography':'France',
    'Age':40, 
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000, 
      

}
```

```python
df=pd.DataFrame([input_data])
```

```python
df
```

```python
df['Gender']=label_encoder.transform(df['Gender'])
```

```python
df
```

```python
ohe=ohe_encoder.transform(df[['Geography']])
ohe.toarray()
ohe=pd.DataFrame(ohe.toarray(),columns=ohe_encoder.get_feature_names_out(['Geography']))
df=pd.concat([df.drop(['Geography'],axis=1),ohe],axis=1)
```

```python
df
```

```python
input_scaled=scaler.transform(df)
```

```python
input_scaled
```

```python
prediction=model.predict(input_scaled)
```

```python
prob=prediction[0][0]
prob
```

```python
if prob>0.5:
    print('customer will exit')

else:
    print('customer will not exit')    
```
